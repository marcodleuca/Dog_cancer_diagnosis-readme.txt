```python

#opening human fasta file 
# Path to your FASTA file
fasta_file ='Homo_sapiens.GRCh38.pep.all.fa'

# Parsing the FASTA file
records = []
for record in SeqIO.parse(fasta_file, "fasta"):
    records.append({
        "ID": record.id,
        "Description": record.description,
        "Sequence": str(record.seq)
    })

# Creating a DataFrame
human= pd.DataFrame(records)


```

```python
#extrapolating information from the desciption and filtering columns for core information
human1=human['Description'].str.split(" ",expand=True)
human1=human1[[0,3,4,7]]
human1.columns=['ID','Gene Ensembl','Transcript Ensembl','Gene Name']
df_human=pd.merge(human, human1, on='ID', how='inner')
df_human['Gene Name']=df_human['Gene Name'].str.replace('gene_symbol:','')
df_human['Gene Ensembl']=df_human['Gene Ensembl'].str.replace('gene:','')
df_human['Transcript Ensembl']=df_human['Transcript Ensembl'].str.replace('transcript:','')
df_human=df_human.drop(columns='Description',axis=1)
```

```python
#assinig to cosmic genes their corrispondent sequence from the fasta file using Gene Name and Transcript
df_cosmic['GENE_SYMBOL'] = df_cosmic['GENE_SYMBOL'].astype(str)
df_cosmic['TRANSCRIPT_ACCESSION'] = df_cosmic['TRANSCRIPT_ACCESSION'].astype(str)
df_human['Gene Name'] = df_human['Gene Name'].astype(str)
df_human['Transcript Ensembl'] = df_human['Transcript Ensembl'].astype(str)


df_cosmic['unique_key'] = df_cosmic['GENE_SYMBOL'] + df_cosmic['TRANSCRIPT_ACCESSION']
df_human['unique_key'] = df_human['Gene Name'] + df_human['Transcript Ensembl']


mapping_dict = df_human.set_index('unique_key')['Sequence'].to_dict()


df_cosmic['SEQUENCE'] = df_cosmic['unique_key'].map(mapping_dict)


df_cosmic = df_cosmic.drop('unique_key', axis=1)


```

```python
df_cosmic
```

```python
#double checking if the position of the AA mutated is corresponding in the sequence assigned 
df_cosmic['AA_POSITION'] = df_cosmic['AA_POSITION'].astype(int)
df_cosmic['MUTATION_AA'] = df_cosmic['MUTATION_AA'].astype(str)
df_cosmic['SEQUENCE'] = df_cosmic['SEQUENCE'].astype(str)


def check_mutation(row):
    position = row['AA_POSITION']-1 
    if position < len(row['SEQUENCE']):
        return row['SEQUENCE'][position] == row['MUTATION_AA']
    else:
        return False  

df_cosmic['HUMAN_MUTATION_MATCH'] = df_cosmic.apply(check_mutation, axis=1)


```

```python
df_cosmic
```

```python
#quantification of positive and negative results on the double check
df_cosmic['HUMAN_MUTATION_MATCH'].value_counts()
```

```python
#checking which letter is in the sequence at the corresponding position for both positive and negative results
df_cosmic['AA_POSITION'] = df_cosmic['AA_POSITION'].astype(int)


def get_letter_at_position(sequence, position):
    
    if position <= len(sequence) and position > 0:
        return sequence[position - 1]
    else:
        return None  


df_cosmic['LETTER_AT_POSITION'] = df_cosmic.apply(lambda row: get_letter_at_position(row['SEQUENCE'], row['AA_POSITION']), axis=1)



```
